```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
party_stats = pd.read_csv('party_wise_results.csv')
party_stats.head(10)
```

Total Parties Contesting in Elections

```python
print(f'{party_stats.shape[0]} Parties contested in Lok - Sabha Elections 2024')
```

Top Parties

```python
top_parties = party_stats.sort_values(by='won', ascending=False).head(5)
top_parties
```

```python
fig = plt.figure(figsize=(10, 5))
plt.bar(top_parties['name'], top_parties['won'], color='crimson')
plt.xlabel = 'party'
plt.ylable = 'seats won'
plt.title('Top 5 parties and seats won')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Parties with Least seats Won (cutoff = 5)

```python
cut_off = 5
parties_with_least_seats = party_stats[ party_stats['won'] <= cut_off ]
parties_with_least_seats[['name', 'won']]
```

Final summary of insights

```python
party_1 = top_parties['name'][0]
party_2 = top_parties['name'][1]
party_3 = top_parties['name'][2]

party_vote_1 = top_parties['won'][0]
party_vote_2 = top_parties['won'][1]
party_vote_3 = top_parties['won'][2]

vote_1_perc = top_parties['won'][0] * 100 / party_stats['won'].sum() 
vote_2_perc = top_parties['won'][1] * 100 / party_stats['won'].sum()
vote_3_perc = top_parties['won'][2] * 100 / party_stats['won'].sum()  

comb_5_vote_perc = top_parties['won'].sum() * 100 / party_stats['won'].sum()
vote12_diff = top_parties['won'][0] - top_parties['won'][1]

cnt_parties_cutoff = parties_with_least_seats.shape[0]
cnt_parties_votes_half = party_stats[ party_stats['won'] >= (party_stats['won'].sum() / 2) ].shape[0]
top_10_parties_perc = party_stats.sort_values(by='won', ascending=False).head(10)['won'].sum() * 100 / party_stats['won'].sum() 


print(f'1. {party_1} has won highest seats among all and holds {vote_1_perc:.2f} % of total votes.', end="\n\n")
print(f'2. {party_2} has won second highest seats among all and holds {vote_2_perc:.2f} % of total votes.', end="\n\n")
print(f'2. {party_3} has won second highest seats among all and holds {vote_3_perc:.2f} % of total votes.', end="\n\n")
print(f'3. Top 5 parties with highest seasts won : ', end="\n\n")
for party in top_parties['name']:
    print(f'\t{party}')
print(f'\nCombined hold seats percentage of {comb_5_vote_perc :.2f} %', end="\n\n")
print(f'4. There is a difference of {vote12_diff} seats between top 2 parties with highest seats. \n indicating that {party_1} has {vote12_diff * 100 / party_vote_1} % more seats than {party_2}', end="\n\n")
print(f'5. No party is currently leading in any seats, the lead column is 0 for all parties.', end="\n\n")
print(f'6. There are {cnt_parties_cutoff} parties with less or equal to 5 seats, indicating a highly regional and smaller party.', end="\n\n")
print(f'7. There are {cnt_parties_votes_half} parties that won 50 % of total seats among all parties.', end="\n\n")
print(f'8. {party_2} struggles to compete with dominating party {party_1} with {vote_1_perc - vote_2_perc :.2f} % less seats.', end="\n\n")
print(f'9. The top 10 parties together hold {top_10_parties_perc:.2f}% of the seats, indicating concentration of power among few top parties.', end="\n\n")
```
